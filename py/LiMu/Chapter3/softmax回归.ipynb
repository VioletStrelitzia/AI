# 要处理的问题类型

## 回归问题

估计一个连续值

- 输出：单连续数值
- 值域：自然区间 $R$
- 损失：跟真实值的区别

## 分类问题

预测一个离散类类别

- 输出：通常多个输出
- 值域：输出置信度
- 损失：跟真实值的区别

### 分类问题示例：

1. MNIST：手写数字识别（10 类）
2. ImageNet：自然物体分类（1000 类）
3. Kaggel上的分类问题：
   - 将人类蛋白质显微镜图分为 28 类
   - 将恶意软件分成 9 类
   - 将恶意的 Wikipedia 评论分为 7 类

### 对类别的编码（encode）

1. 一位有效编码 $$\boldsymbol y = {[y_1, y_2,\dots,y_n]}^T$$
$$
f(x)=\left\{
\begin{aligned}
1  &      & {\text{if } i = y}\\
0  &      & {\text{otherwise}}\\
\end{aligned}
\right.
$$
   - 使用均方损失
   - 最大值作为预测 $$\hat{y} = \mathop{argmax}\limits_{i}{o_i}$$
   - 无校验比例的情况下，比较置信度

### softmax

$$\hat{y}_{i} = \frac{\exp{o_i}}{\sum_{k}{\exp{o_k}}}$$

- 使用 e 指数：保证非负

### 交叉熵损失

- 常用于衡量两个概率的区别

- 对于真实概率 $\boldsymbol y$ 和预测概率 $\hat{\boldsymbol y}$:

$$l(\boldsymbol y, \hat{\boldsymbol y}) = - \sum_{i}{\hat{y_i \log{y_i}}}$$

- 其梯度是真实概率和预测概率的区别

$$\partial_{o_i}l(\boldsymbol y, \hat{\boldsymbol y}) = softmax(\boldsymbol o)_i - y_i$$
