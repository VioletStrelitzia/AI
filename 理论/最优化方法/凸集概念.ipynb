##### 一条直线

已知两个点（向量表示） $\boldsymbol{x}_1, \boldsymbol{x}_2$，经过两点的直线为：

$$
\boldsymbol{x} = \theta \boldsymbol{x}_1 + (1-\theta) \boldsymbol{x}_2 ,\quad \theta \in \boldsymbol{R}
$$

#### 组合

##### 仿射组合

- n 个向量的仿射组合为：

$$
\boldsymbol{x} = \sum^{n}_{i=1}\theta_i \boldsymbol{x}_i ,\quad \sum^{n}_{i=1}\theta_i = 1
$$

- 满足 $\sum^{n}_{i=1}\theta_i = 1$，就是一个仿射组合。
- 是一条直线。

##### 凸组合

- n 个向量的凸组合为：

$$
\boldsymbol{x} = \sum^{n}_{i=1}\theta_i \boldsymbol{x}_i ,\quad \sum^{n}_{i=1}\theta_i = 1 \quad and \quad \forall \theta_i \geq 0
$$

- 满足 $\sum^{n}_{i=1}\theta_i = 1 \quad and \quad \forall \theta_i \geq 0$，就是一个凸组合。
- 是一条线段。

##### 锥组合

- n 个向量的锥组合为：

$$
\boldsymbol{x} = \sum^{n}_{i=1}\theta_i \boldsymbol{x}_i ,\quad \forall \theta_i \geq 0
$$

- 满足 $\forall \theta_i \geq 0$，就是一个锥组合。
- 是一条射线。

#### 集合

##### 仿射集合

- 仿射集中的任意两个点连成的直线上的都在集合 $\boldsymbol{C}$ 内。
- 等价于
$$
\forall \boldsymbol{x}_i \in \boldsymbol{C} \\
\boldsymbol{x} = \sum^{n}_{i=1}\theta_i \boldsymbol{x}_i \in \boldsymbol{C} ,\quad \sum^{n}_{i=1}\theta_i = 1
$$

<div style="border: 1px solid #e1e4e8; padding: 10px; margin: 10px 0; border-radius: 6px;">
证明:
由数学归纳法，只需要证明从 n 个点到 n + 1 个点的仿射组合都在集合 $\boldsymbol{C}$ 内。

n = 2 时，显然成立。下文默认 $\theta$ 和均为1。

要证 n = k+1 时也成立，即要证：
$$
\sum^{k+1}_{i=1}\theta_i \boldsymbol{x}_i \in \boldsymbol{C}
$$

构造：
$$
x'=\sum^{k}_{i=1}{\frac{\theta_i}{\sum^{k}_{j=1}\theta_j}\boldsymbol{x}_i} \in \boldsymbol{C}
$$

n = 2 时成立，则：
$$
(\sum^{k}_{j=1}\theta_j)x' + (1-\sum^{k}_{j=1}\theta_j)x_{k+1} \in \boldsymbol{C} \\
$$
</div>

- 仿射集是 $\boldsymbol{R}^n$ 整个空间。

##### 凸集

- 仿射集中的任意两个点连成的直线上的都在集合 $\boldsymbol{C}$ 内。
- 凸多边形。


#### 子空间

##### 仿射子空间
- 条件
  - 包含原点。
  - 对数乘和加法封闭。
- 定义
  - 设 $\boldsymbol{C}$ 是一个仿射集，$\{x-x_0|x,x_0 \in \boldsymbol{C}\}$ 是其子空间。

#### 包

##### 仿射包

增加最少的点，使得集合成为一个仿射集。

#### 线性方程组的解

线性方程组：$\boldsymbol{Ax}=\boldsymbol{b}$ 的解空间 $\boldsymbol{C}$ 是一个仿射集。

<div style="border: 1px solid #e1e4e8; padding: 10px; margin: 10px 0; border-radius: 6px;">
证明：

无解时，$\emptyset$ 是一个仿射集。

有解时，
1. 只有一个解时，解空间只有一个点，是一个仿射集。
2. 有多个解时，只需证：
$$
\theta \boldsymbol{x}_1 + (1-\theta) \boldsymbol{x}_2 \in \boldsymbol{C} ,\quad \forall \boldsymbol{x}_1, \boldsymbol{x}_2 \in \boldsymbol{C} \quad and \quad \theta \in \boldsymbol{R} 
$$
左乘 $\boldsymbol{A}$ 得：
$$
\begin{aligned}
\boldsymbol{A}(\theta \boldsymbol{x}_1 + (1-\theta) \boldsymbol{x}_2) =& \theta\boldsymbol{A}\boldsymbol{x}_1 + (1-\theta)\boldsymbol{A}\boldsymbol{x}_2 \\
=& \theta\boldsymbol{b} + (1-\theta)\boldsymbol{b} \\
=& \boldsymbol{b}
\end{aligned}
$$
则 $\theta \boldsymbol{x}_1 + (1-\theta) \boldsymbol{x}_2$ 也是方程组的解，得证。

</div>
